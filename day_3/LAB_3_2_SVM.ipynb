```python
import pandas as pd
```

```python
# Load the sepal data
df = pd.read_csv('data/sepal_data.csv')
X = df.iloc[:, :3]
y = df.iloc[:, 4]
df.head()
```

```python
# Plot the data (Sepal length by sepal width)
import matplotlib.pyplot as plt

plt.figure(figsize = (6, 4))
plt.scatter(X['sepal_length'], X['sepal_width'], cmap = 'viridis', c = y)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Width V Length')
plt.plot
```

```python
# Split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2)
```

```python
# Make & train the SVM
from sklearn.svm import SVC
model = SVC(kernel = 'rbf')
model.fit(X_train, y_train)
```

```python
# Run the model on the testing and training data
y_pred = model.predict(X_test)
y_pred_train = model.predict(X_train)
```

```python
# Get the accuracy score
from sklearn.metrics import accuracy_score, classification_report
print(accuracy_score(y_test, y_pred))
print(accuracy_score(y_train, y_pred_train))
```

```python
# Run the classification report
print(classification_report(y_train, y_pred_train))
print(classification_report(y_test, y_pred))
```
